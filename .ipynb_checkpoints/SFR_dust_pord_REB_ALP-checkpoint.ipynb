```python
#my functions
from scipy.integrate import quad
from matplotlib import cm
col_f = cm.get_cmap('viridis')
col_f1=cm.get_cmap('viridis')
from librerie import *
fold_in = '../'
from scipy.optimize import fsolve
from astropy.cosmology import Planck15 as cosmo
from matplotlib import gridspec
from scipy.optimize import curve_fit
from scipy import stats


####### useful constants
alphab=2.6e-13
G = 6.67e-8#cgs
bi = -2.5 #UV spectral slope for MW
yr=3.154e+7#in s
#p0=numpy.array([10**4.0,10**8.0])
Msol=1.99*1e33#g
Lsol=3.9*1e33#erg/s
mp=1.66e-24#g, proton mass
kb = 1.38064852e-16#cgs
h = 6.63e-27 #ergs
c = 3e10 #cm/s
pc=3.09e18#cm
To = 2.725#TCMB
Zsole=0.0142
D=1.0/162.0#dust to gas ratio MW


################## MY FUNCTIONS #########################################################################################################
def log_MD_MAX(logMstar,yd):
        #SLPETER IMF 1-100
        return logMstar+np.log10(yd)-np.log10(53.0)


def gzg6(redshift):
        
        gz=(1.0+redshift)/(cosmo.luminosity_distance(redshift).value)**2.0
        g6=7.0/(cosmo.luminosity_distance(6.0).value)**2.0
        #print g6
        return gz/g6

def err_estimate_arr(array):
        #median, - err, + err
        return np.array([numpy.median(array), (-numpy.percentile(array,16)+numpy.median(array)),(numpy.percentile(array,84)-numpy.median(array))]) 

def alpha_andrew(Sigma_SFR,ks,y):#,r_CII  
        #ks_mean is the mean deviation of the galaxy sample form KS relation
        #SFR_array in Msol/(yr)
        return 32.47*(y**2)*(Sigma_SFR**(-2.0/7.0))/(ks**(5.0/7.0))
        
def freq_convert(lamda):
        #converte lamda in micron in frequenze in GHz
        return c*1e-5/lamda

def temp_convert(lamda):
        #converte in T numeratore planck function da lamda in micron
        nu=freq_convert(lamda)
        return h*nu*1e9/kb

def kabs_l(lamda,bd):
        return kabs*(lamda/lamda_abs)**-bd

def T_CMB(redshift):
        return To*(1+redshift)
        
def Bnu(T_obs,T):
        return 1.0/(numpy.exp(T_obs/T)-1.0)
# objective function
def lin_fit(x, a, b):
	return a * x + b

#correction to temperature due to CMB heating
def T_CMB_corr(Td,redshift,bd):
        return (Td**(4+bd)+(To**(4+bd))*(((1.0+redshift)**(4+bd))-1.0))**(1.0/(4.0+bd))

def flusso_obs(lamda,Td1,logMdust,redshift,bd):        
        DL=cosmo.luminosity_distance(redshift).value
        nu=freq_convert(lamda)*1e9#Hz
        T_obs=temp_convert(lamda)
        k_obs=kabs_l(lamda,bd)#cm**2/g 
        const=(2*h*(nu**3)/c**2)*Msol*1e26/(pc*1e6)**2.0#nb: fattore moltiplicativo avanti dato da const*Msol[g]*milliJscky/Mpc**2
        return (const*(1.0+redshift)*(10**logMdust)*k_obs*(Bnu(T_obs,T_CMB_corr(Td1,redshift,bd))-Bnu(T_obs,T_CMB(redshift)))/DL**2)

def flusso_obs_funct(lamda,Td1,logMdust,redshift):
        bd=2.03 #commenta se voglio fittare anche su bd
        DL=cosmo.luminosity_distance(redshift).value
        nu=freq_convert(lamda)*1e9#Hz
        T_obs=temp_convert(lamda)
        k_obs=kabs_l(lamda,bd)#cm**2/g 
        const=(2*h*(nu**3)/c**2)*Msol*1e26/(pc*1e6)**2.0#nb: fattore moltiplicativo avanti dato da const*Msol[g]*milliJscky/Mpc**2
        #print const
        return (const*(1.0+redshift)*(10**logMdust)*k_obs*(Bnu(T_obs,T_CMB_corr(Td1,redshift,bd))-Bnu(T_obs,T_CMB(redshift)))/DL**2)

def TdCII_anal_funct(lamda,redshift,Flux,LCII,alphaCII,metallicity):#micron, mJy, Lsol, Msol/Lsol, Zsol 
        T_obs=temp_convert(lamda)
        Ftilde=(0.98e-16)*Flux#mJy
        A=(4.33e-24)*gzg6(redshift)*LCII*alphaCII*D*metallicity/(cost_fd)
        f=Bnu(T_obs,T_CMB(redshift))+Ftilde/A
        TdCII=T_obs/numpy.log( 1.0+ (f**-1.0) )
        return TdCII

def logL_IR(logMdust,bd,Td):
        nu158=1e9*freq_convert(158)
        Teta=(np.pi*8.0/c**2) * (kabs/nu158**bd) * ((kb**(4.0+bd))/(h**(3.0+bd))) * scipy.special.zeta(4.0+bd) * scipy.special.gamma(4.0+bd)
        #print 'Teta MIA da AW dust model:', (Msol*Teta/Lsol)**(-1.0/(4.0+bd))
        #print 'prod', scipy.special.zeta(4.0+bd) * scipy.special.gamma(4.0+bd)
        return logMdust+np.log10(Teta)+(4.0+bd)*np.log10(Td)+np.log10(Msol/Lsol)

##### dust model#############
#bd = numpy.abs(np.random.uniform(0.5,3.5,n))
bd=2.03
kabs=10.41
lamda_abs=158.0
```
